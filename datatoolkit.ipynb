```python
'''
Q4.  How do you calculate the correlation matrix using Seaborn and visualize
it with a heatmap
'''
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = {
    'A': [10, 20, 30, 40, 50],
    'B': [5, 15, 25, 35, 45],
    'C': [2, 4, 6, 8, 10],
    'D': [3, 6, 9, 12, 15]
}

df = pd.DataFrame(data)


corr_matrix = df.corr()

plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title("Correlation Matrix Heatmap")

plt.show()
```

```python
'''
Q5.  Generate a bar plot using Plotly
'''
import plotly.express as px
import pandas as pd

data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': [10, 25, 15, 30, 20]
}

df = pd.DataFrame(data)

fig = px.bar(df, x='Category', y='Values', title="Bar Plot Example",
             color='Values', text='Values', labels={'Values': 'Count'})

fig.show()
```

```python
'''
Q8. Create a line plot with multiple lines using Matplotlib
'''
import matplotlib.pyplot as plt
#  Data points of line 1
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
# Data points of line 2
x2 = [1, 2, 3, 4, 5]
y2 = [1, 3, 5, 7, 9]
# Data points of line 3
x3 = [1, 2, 3, 4, 5]
y3 = [5, 4, 3, 2, 1]
# Plotting all lines with specifying labels
plt.plot(x1, y1, label='Line 1')
plt.plot(x2, y2, label='Line 2')
plt.plot(x3, y3, label='Line 3')
# Adding legend, x and y labels, and titles for the lines
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plotting Multiple Lines')
# Displaying the plot
plt.show()
```

```python
'''
Q9. Generate a Pandas DataFrame and filter rows where a column value is
greater than a threshold
'''
import pandas as pd

data = {
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Age': [25, 30, 35, 40, 28]
}

df = pd.DataFrame(data)

threshold = 30

filtered_df = df[df['Age'] > threshold]

print(filtered_df)
```

```python
'''
Q11. Perform matrix multiplication using NumPy
'''
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

result = A @ B  # OR use np.dot(A, B)

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix Multiplication Result:\n", result)
```
